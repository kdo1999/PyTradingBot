```python
# DataFrame 만드는 방법
import pandas as pd

df = pd.DataFrame()
df["Name"] = ["jacky", "Steven", "George"]
df["Age"] = [38, 25, 23]
df["Driver"] = [True, False, True]

df
```

```python
# python dictionary를 사용하여 DataFrame 만드는 방법
data = {
    "Name":["Jacky", "Steven", "George"],
    "Age": [38, 25, 23],
    "Driver": [True, False, True]
}

df = pd.DataFrame(data)
df
```
