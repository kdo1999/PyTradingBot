## DataFrame 인덱싱
- 자료구조에서 하나의 값에 접근하는 것을 인덱싱이라고 합니다.
- Pandas

```python
# python dictionary를 사용하여 DataFrame 만드는 방법
data = {
    "Name":["Jacky", "Steven", "George", "Harry"],
    "Age": [38, 25, 23, 10],
    "Driver": [True, False, True, True]
}

df = pd.DataFrame(data)
df.index = ["a", "b", "c", "d"]
df.index.name = "index"

df
```
